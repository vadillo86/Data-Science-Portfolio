 # Regresion Linealde prediccion de precios de casas
 
 El objetivo de este ejercicio es el de obtener una predicción de los precios de las casas, conocidas unas características de las mismas y su precio.
 
 + Sobre el dataset que utilizaremos se tienen las siguientes características:
 
    1. pies_cuadrados_casa
    2. num_baños
    3. num_habitaciones
    4. num_cocinas
    5. num_coches_garaje
    6. pies_cuadrados_garaje
    7. precio
    
+ Se realizará una regresión lineal simple, tomando como *variable independiente los pies cuadrados* de la casa y como *variable dependiente el precio*.

+ En este ejercicio se mostrará lo siguiente:

    1. Carga de datos (Resuelto)
    2. *Cambio de estructura de datos a numpy* para el uso de Scikit-Learn
    3. *Visualizar con una gráfica de dispersión la relación entre precio y pies cuadrados de las casas*
    4. *Creacción del modelo y ajuste del mismo*
    5. *Obtención del modelo* (Resta de Regresión)
    6. *Visualización del modelo* en el que se muestre una gráfica de dispersión con la elación entre precio y pies cuadrados de las casas y por otro lado (y en la misma gráfica) la recta de regresión.
    7. *Medición de la calidad de los resultados*: Obtención del Coeficiente de Determinación R2

## Importar librerías

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline 
```

## 1. Carga del dataset

```python
df = pd.read_csv('../data/precio_casas.csv', header=0, sep=';')
df.head()
```

## 2. Cambio de estructura de datos a numpy

```python
X = df['pies_cuadrados_casa'].values
y = df['precio'].values

X = np.reshape(np.asarray(X), (len(X), 1))
```

## 3. Visualizar con una gráfica de dispersión la relación entre precio y pies cuadrados de las casa

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.scatter(X, y,  color='green', marker='.', )
ax.set_title("Precio de casas por pies cuadrados")
ax.set_xlabel("Pies cuadrados")
ax.set_ylabel("Precio")
plt.show()
```

## 4. Creacción del modelo y ajuste del mismo

```python
reg = linear_model.LinearRegression()

reg.fit(X,y)
```

## 5. Obtención del modelo (Resta de Regresión)

```python
a = reg.coef_[0]
b = reg.intercept_

print ("Precio = {a:0.1f} · Pies_cuadrados + {b:0.1f}".format(a=a, b=b))
```

## 6. Visualización del modelo en el que se muestre una gráfica de dispersión con la relación entre precio y pies cuadrados de las casas y por otro lado (y en la misma gráfica) la recta de regresión.

```python
predicciones = reg.predict(X)

fig, ax = plt.subplots(figsize=(8,6))
ax.scatter(X, y,  color='green', marker='.', )
ax.plot(X, predicciones, color='red')
ax.set_title("Precio de casas por pies cuadrados")
ax.set_xlabel("Pies cuadrados")
ax.set_ylabel("Precio")
plt.show()
```

## 7. Medición de la calidad de los resultados: Obtención del Coeficiente de Determinación R2

```python
r2 = reg.score(X, y)

print ("Coeficiente de determinación R^2: {r2}".format(r2=r2))
```
